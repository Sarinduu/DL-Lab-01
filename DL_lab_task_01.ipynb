2. Create a random 100000x1 array from the same exponential distribution and visualize the
distribution histogram along with uniform and normal distributions. Use the code given
below and do the necessary changes. Change the number of bins and the plot view to make
the visualization better.

```python
scale = 1.0

exp_data = np.random.exponential(scale, 100000)

uniform_data = np.random.uniform(0, 1, 100000)
normal_data = np.random.normal(0, 1, 100000)

plt.figure(figsize=(12, 8))
bins = 100

plt.hist(exp_data, bins=bins, alpha=0.6, color='b', label='Exponential')

plt.hist(uniform_data, bins=bins, alpha=0.6, color='r', label='Uniform')

plt.hist(normal_data, bins=bins, alpha=0.6, color='g', label='Normal')

plt.title('Distributions Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend(loc='upper right')

plt.show()
```

3. Using matplotlib, plot the Z = X**2 + Y**2 in a 3D plot. Limit the x and y values to [-5, 5].

```python
x = np.linspace(-5, 5)
y = np.linspace(-5, 5)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2
figure = plt.figure(1, figsize = (12, 4))
subplot3d = plt.subplot(111, projection='3d')
surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0.1)
plt.show()
```

4. Calculate the pearson standard coefficient and spearman rank correlation for the [HP,
Attack, Defense, Sp. Ark, Sp. Def, Speed] features in the seaborn tutorial.ipynb. Visualize the
results using heatmaps. Make sure the values are displaced in the heatmap as well.

```python
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv

df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
data = df[features]

pearson_corr = data.corr(method='pearson')

spearman_corr, _ = spearmanr(data)

spearman_corr_df = pd.DataFrame(spearman_corr, index=features, columns=features)

plt.figure(figsize=(10, 8))
sns.heatmap(pearson_corr, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Pearson Correlation Heatmap')
plt.show()

plt.figure(figsize=(10, 8))
sns.heatmap(spearman_corr_df, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Spearman Correlation Heatmap')
plt.show()
```
